Split up the votes dictionary into useful, funny and cool; add a column for each value to the data frame

```python
list1 = []
list2 = []
list3 = []
for i in df.votes:
    list1.append(i['useful'])
    list2.append(i['funny'])
    list3.append(i['cool'])
df['useful'] = pd.Series(list1, index=df.index)
df['funny'] = pd.Series(list2, index=df.index)
df['cool'] = pd.Series(list3, index=df.index)
```

Count the number of words in text

```python
list4 = []
for i in df.text:
    list4.append(len(i.split()))

df['word_nr'] = pd.Series(list4, index=df.index)
```

Tokenize the text for each review

```python
import nltk
from nltk.tokenize import RegexpTokenizer
from nltk.corpus import stopwords

stopset = set(stopwords.words('english'))
tokenizer = RegexpTokenizer(r'\w+')

list5=[]
for each in df.text:
    tokens = tokenizer.tokenize(each)
    doc = [word for word in tokens if word not in stopset and len(word) > 3]
    list5.append(doc)
words_list=[]
for wlist in list5:
    words=[]
    for word in wlist:
        words.append(word.lower())
    words_list.append(words)
df['word_list'] = pd.Series(words_list, index=df.index)
#df
```

```python
list6 = []
for i in df.word_list:
    list6.append(len(i))

df['word_nr_rel'] = pd.Series(list6, index=df.index)
df['ratio_noStop']=df.word_nr_rel/df.word_nr
#pd.scatter_matrix(df);
```

```python
df_sorted=df.sort('useful',ascending=False)
df_50=df_sorted[:50]
df_50=df_50[['useful','funny','cool','word_nr','word_list','word_nr_rel']]
#df_50
```

```python
type_list=[]
for wlist in df.word_list:
    type_list.append(nltk.pos_tag(wlist))
df['word_list_type']= pd.Series(type_list, index=df.index)
```

```python
df['word_list_type'][1][1][1]
list1=[]
list2=[]
list3=[]
for each in df['word_list_type']:
    for every in each:
        list1.append(every[1])
    list2.append(list1)
df['word_type'] = pd.Series(list2, index=df.index)
```

```python
from nltk.corpus import sentiwordnet as swn
pos_list=[]
neg_list=[]
obj_list=[]
index = 0
for wlist in df.word_list:
    pos=[]
    neg=[]
    obj=[]
    index2=0
    for word in wlist:
        #if df.word_type[index][index2]=='JJ':
        doc = list(swn.senti_synsets(word))
        try:
            doc0=list(doc)[0]
            pos.append(doc0.pos_score())
            neg.append(doc0.neg_score())
            obj.append(doc0.obj_score())
        except IndexError:       
            pos.append(0)
            neg.append(0)
            obj.append(0)
        index2=index2+1
    pos_list.append(sum(pos)/len(pos))
    neg_list.append(sum(neg)/len(pos))
    obj_list.append(sum(obj)/len(pos))
    index=index+1
df['pos'] = pd.Series(pos_list, index=df.index)
df['neg'] = pd.Series(neg_list, index=df.index)
df['obj'] = pd.Series(obj_list, index=df.index)
```
