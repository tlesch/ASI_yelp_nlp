Split up the votes dictionary into useful, funny and cool; add a column for each value to the data frame

```python
list1 = []
list2 = []
list3 = []
for i in df.votes:
    list1.append(i['useful'])
    list2.append(i['funny'])
    list3.append(i['cool'])
df['useful'] = pd.Series(list1, index=df.index)
df['funny'] = pd.Series(list2, index=df.index)
df['cool'] = pd.Series(list3, index=df.index)
```
